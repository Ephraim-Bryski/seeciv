```python
import matplotlib.pyplot as plt
import numpy as np

def get_max_step(n):

    li = np.round(np.random.random((1,n)))[0]

    flip_idxs = np.where(li[0:-1] != li[1:])[0] 

    both_steps = np.diff(flip_idxs)
    steps = both_steps[::2]

    if np.size(steps) == 0:
        return 0
    else:
        return np.max(steps)


n_trials = 1000

def get_average_max_step(n):

    max_steps = [get_max_step(n) for _ in range(n_trials)]
    return np.mean(max_steps)


tests = [10,20,50,100,500,1000,2000,5000,10000,100_000]#,1_000_000]


average_steps = [get_average_max_step(n) for n in tests]

plt.plot(np.log2(tests),average_steps)
plt.plot(average_steps,average_steps,'g')

plt.xlabel('log2(Number of Digits in Addition)')
plt.ylabel('Average Max Gap Between XOR(A,B)=1')
```
